```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

def plot_region_fpop_sensitivity(df, x_axis, rot_x=True, show_xlabel=True):
    plt.figure(figsize=(18, 6))

    pal = sns.color_palette("Oranges_r", df.shape[0])

    rank = df['sum_flow'].argsort().argsort()
    sns.barplot(x=x_axis, y='sum_flow', data=df, palette=np.array(pal[::-1])[rank])
    plt.axhline(0.35, color='red', alpha=0.5, label='서울시 노원구/중구 코로나19 유동인구 민감율')
    for idx, score in enumerate(df.sum_flow):
        plt.text(idx-0.4, score+0.01, score, size=13, weight='bold')

    plt.title(f'서울시 노원구/중구 {x_axis}별 코로나19 유동인구 민감율', size=25)
    plt.legend(fontsize=15)
    plt.ylabel('코로나19 유동인구 민감율', size=15)
    if rot_x:
        plt.xticks(rotation=45, size=15, ha='right')
    if show_xlabel:
        plt.xlabel(x_axis, size=15)
    else:
        plt.xlabel(None)

    plt.show()
```
